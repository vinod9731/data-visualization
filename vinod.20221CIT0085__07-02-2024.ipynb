```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step 1")
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
x=np.linspace(0,150,100)
ax.plot(x,np.sin(x));
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
x=np.linspace(0,5,10)
ax.plot(x,np.sin(x));
print(x)
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
x=np.linspace(0,5,10)
ax.plot(x,np.sin(x));
print(x)
ax.set_title('simple plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
x=np.linspace(0,5,10)
ax.plot(x,np.sin(x));
ax.plot(x,np.cos(x));
print(x)
ax.set_title('multiple lines');
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
x=np.linspace(0,5,10)
ax.plot(x,np.sin(x),label='sin');
ax.plot(x,np.cos(x),label='cos');
ax.set_title('multiple lines');
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin');
ax.plot(x,np.cos(x),label='cos');
ax.plot(x,np.tan(x),label='tan');
ax.set_title('multiple lines');
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',color ='red');
ax.plot(x,np.cos(x),label='cos',color='g');
ax.plot(x,np.tan(x),label='tan');
ax.set_title('multiple lines');
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',linestyle='dashed');
ax.plot(x,np.cos(x),label='cos',linestyle='dotted');
ax.plot(x,np.sin(x+1),label='cos',linestyle='dashdot');

ax.set_title('multiple lines');
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
plt.show()
```
